Paso 1: Importar librerias necesarias

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Paso 2: Cargar el dataset tips usando seaborn.

```python
df = sns.load_dataset("tips")

print("Dataset 'tips' cargado correctamente.")
print("Filas y columnas:", df.shape)
```

Paso 3: Realizar un análisis preliminar: utiliza .head(), .info(), .describe() para obtener una visión general de los datos.

```python
print("Primeras filas del dataset:")
df.head()

print("Informacion general del dataset:")
df.info()

print("Estadísticas descriptivas (numericas):")
df.describe()
```

Paso 4: Verificar la existencia de valores nulos o datos faltantes.

```python
print("Valores nulos por columna:")
df.isna().sum()
```

a) Mapa de Calor (Heatmap):

* Objetivo: Visualizar la correlación entre las variables numéricas en el dataset.
* Instrucciones: Utiliza un mapa de calor para mostrar la matriz de correlación entre las variables numéricas (total_bill, tip, size).
* Asegúrate de personalizar el mapa con una paleta de colores y etiquetas en los ejes.

Paso extra: Antes que nada definimos un tema y una paleta de colores

```python
sns.set_theme(style="white")
color_pallet = "mako"
```

```python
sns.set_theme(style="white")

numeric_cols = ["total_bill", "tip", "size"]
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap=color_pallet,
    center=0,
    square=True
)

plt.title("Correlacion entre variables numericas (tips)")
plt.xlabel("Variables")
plt.ylabel("Variables")
plt.show()
```

b) Diagrama de Violín:

* Objetivo: Mostrar la distribución de las propinas (tip) para diferentes días de la semana.
* Instrucciones: Utiliza un diagrama de violín para comparar cómo varía la distribución de las propinas en cada uno de los días (day).
* Personaliza el gráfico con colores y etiquetas adecuadas.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(
    data=df,
    x="day",
    y="tip",
    hue="day",
    palette=color_pallet,
    legend=False,
    inner="quartile",
    linewidth=1
)

plt.title("Distribucion de la propina (tip) por dia")
plt.xlabel("Dia")
plt.ylabel("Propina")
plt.show()
```

c) Gráfico de dispersión (Scatter Plot):

* Objetivo: Analizar la relación entre el total de la cuenta (total_bill) y la propina (tip).
* Instrucciones: Crea un gráfico de dispersión con total_bill en el eje X y tip en el eje Y. Usa un color diferente para cada día de la semana.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",
    style="day",
    palette=color_pallet,
    alpha=0.8,
    s=70
)

plt.title("Relacion entre total de la cuenta y propina (por dia)")
plt.xlabel("Total de la cuenta (total_bill)")
plt.ylabel("Propina (tip)")
plt.legend(title="Dia")
plt.show()
```

d) Gráfico de barras (Bar Plot):

* Objetivo: Comparar el promedio de la propina (tip) por género (sex).
* Instrucciones: Utiliza un gráfico de barras para mostrar el promedio de las propinas por género (hombres y mujeres).

```python
plt.figure(figsize=(6, 5))
sns.barplot(
    data=df,
    x="sex",
    y="tip",
    hue="sex",
    palette=color_pallet,
    legend=False,
    errorbar=("ci", 95),
    linewidth=1
)

plt.title("Promedio de propina (tip) por genero")
plt.xlabel("Genero")
plt.ylabel("Propina promedio")
plt.show()
```
