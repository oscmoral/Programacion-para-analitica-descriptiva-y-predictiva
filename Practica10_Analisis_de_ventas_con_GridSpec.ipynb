**Practica: Análisis de Ventas con GridSpec**

Objetivo:
Utilizar GridSpec para organizar subplots de diferentes tamaños en un análisis de ventas.

1. Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros. Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.
2. Usa GridSpec para organizar los subplots de la siguiente manera:
* Primer gráfico (barra, ocupando dos columnas)
* Segundo gráfico (línea, una columna)
* Tercer gráfico (dispersión, dos filas).
* Agrega títulos y ajusta espacios.

Paso 1: Importar librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

Paso 2: Genera un DataFrame de ventas con Mes, Ventas, y Gastos. Utiliza randint para generar números aleatorios enteros. Para ventas los valores deben estar entre 2000 y 5000. Para gastos, los valores debn estar entre 1000 y 4000.

```python
np.random.seed(42)

months = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

sales = np.random.randint(2000, 5001, size=len(months))
expenses = np.random.randint(1000, 4001, size=len(months))

df = pd.DataFrame({
    "Mes": months,
    "Ventas": sales,
    "Gastos": expenses
})

df.head()
```

Paso 3: Usa GridSpec para organizar los subplots de la siguiente manera:
* Primer gráfico (barra, ocupando dos columnas)
* Segundo gráfico (línea, una columna)
* Tercer gráfico (dispersión, dos filas).
* Agrega títulos y ajusta espacios.

```python
fig = plt.figure(figsize=(12, 7))
fig.subplots_adjust(hspace=0.35, wspace=0.25)

gs = GridSpec(2, 3, figure=fig)

ax_bar = fig.add_subplot(gs[0, 0:2])
ax_bar.bar(df["Mes"], df["Ventas"])
ax_bar.set_title("Ventas por mes (barras)")
ax_bar.set_xlabel("Mes")
ax_bar.set_ylabel("Ventas")

ax_line = fig.add_subplot(gs[0, 2])
ax_line.plot(df["Mes"], df["Gastos"], marker="o")
ax_line.set_title("Gastos por mes (linea)")
ax_line.set_xlabel("Mes")
ax_line.set_ylabel("Gastos")

ax_scatter = fig.add_subplot(gs[1, :])
ax_scatter.scatter(df["Ventas"], df["Gastos"])
ax_scatter.set_title("Relacion entre ventas y gastos (dispersion)")
ax_scatter.set_xlabel("Ventas")
ax_scatter.set_ylabel("Gastos")

plt.tight_layout()
plt.show()
```
