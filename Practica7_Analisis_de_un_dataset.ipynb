Paso 1: Importar librerias necesarias

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Paso 2: Cargar el dataset Iris utilizando seaborn o cualquier otro dataset de su elección.

```python
df = sns.load_dataset('iris')

print("Dataset Iris cargado correctamente.")
print("Filas y columnas:", df.shape)
```

Paso 3: Verificar las primeras filas del dataset con el método .head().

```python
print("Primeras filas del dataset:")
df.head()
```

Paso 4: Obtener información general sobre los datos (nombres de columnas, tipos de datos, valores nulos, etc.).

```python
print("Informacion general:")
df.info()

print("\nValores nulos por columna:")
print(df.isna().sum())

print("\nNombres de columnas:")
print(df.columns.tolist())
```

a) Gráfico de barras:

* Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.
* Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
group_means = df.groupby("species")[["petal_length", "petal_width"]].mean()

x = np.arange(len(group_means.index))
bar_width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - bar_width/2, group_means["petal_length"], width=bar_width, label="Largo del pétalo")
plt.bar(x + bar_width/2, group_means["petal_width"], width=bar_width, label="Ancho del pétalo")

plt.xticks(x, group_means.index)
plt.title("Promedio del largo y ancho del petalo por especie")
plt.xlabel("Especie")
plt.ylabel("Promedio (cm)")
plt.legend()
plt.show()
```

b) Histograma:

* Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
* Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
plt.figure(figsize=(8, 5))
plt.hist(df["sepal_length"], bins=12)
plt.title("Distribucion del largo del sepalo (sepal_length)")
plt.xlabel("Largo del sepalo (cm)")
plt.ylabel("Frecuencia")
plt.show()
```

c) Gráfico de dispersión (scatter plot):

* Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.
* Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
plt.figure(figsize=(8, 5))

species_list = df["species"].unique()

for sp in species_list:
    subset = df[df["species"] == sp]
    plt.scatter(subset["petal_length"], subset["petal_width"], label=sp)

plt.title("Relacion entre largo y ancho del petalo")
plt.xlabel("Largo del petalo (cm)")
plt.ylabel("Ancho del petalo (cm)")
plt.legend(title="Especie")
plt.show()
```

d) Gráfico de cajas (box plot):

* Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
* Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.

Creamos un grafico para el largo del sepalo por especie:

```python
plt.figure(figsize=(8, 5))

data = [df[df["species"] == sp]["sepal_length"] for sp in df["species"].unique()]
plt.boxplot(data, tick_labels=df["species"].unique())

plt.title("Distribucion del largo del sepalo por especie")
plt.xlabel("Especie")
plt.ylabel("Largo del sépalo (cm)")
plt.show()
```

Creamos un grafico para el ancho del sepalo por especie:

```python
plt.figure(figsize=(8, 5))

data = [df[df["species"] == sp]["sepal_width"] for sp in df["species"].unique()]
plt.boxplot(data, tick_labels=df["species"].unique())

plt.title("Distribucion del ancho del sepalo por especie")
plt.xlabel("Especie")
plt.ylabel("Ancho del sépalo (cm)")
plt.show()
```

e) Gráfico de líneas:

* Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.
* Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
plt.figure(figsize=(10, 4))
plt.plot(df.index, df["petal_length"])

plt.title("Tendencia del largo del petalo (petal_length) a traves del dataset")
plt.xlabel("Indice de la observacion")
plt.ylabel("Largo del petalo (cm)")
plt.show()
```
