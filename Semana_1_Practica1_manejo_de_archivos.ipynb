# 3.- Se recorre el archivo



1.   Se declara una lista vacia.
2.   Abrimos el archivo y recorremos el archivo por linea.
3.   Utilizamos la funcion startswith para filtrar la informacion por "X-DSPAM-Confidence:".
4.   Guardamos el valor numerico en la lista



```python
values = []

with open(path, "r", encoding="utf-8", errors="ignore") as archivo:
    for line in archivo:
        if line.startswith("X-DSPAM-Confidence:"):
            numero_str = line.split(":")[1].strip()
            numero = float(numero_str)
            values.append(numero)
```

Despues imprimimos los valores de la lista

```python
print("Mediciones encontradas:")
for v in values:
    print(v)
```

En el siguiente paso se realizan los calculos necesarios

```python
n = len(values)
total = sum(values)
average = total / n if n > 0 else 0

variance = sum((x - average) ** 2 for x in values) / n if n > 0 else 0
```

Por ultimo imprimimos los resultados

```python
print("\n--- Resultados ---")
print("Cantidad de datos extraídos:", n)
print("Sumatoria:", total)
print("Promedio:", average)
print("Varianza (poblacional):", variance)
```
