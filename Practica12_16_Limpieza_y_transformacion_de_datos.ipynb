## PrÃ¡ctica 14: NormalizaciÃ³n de los nombres de la columna.

Paso 1: Imprime los tipos de datos de cada columna.

Paso 2: AsegÃºrate que las columnas de SALE PRICE(sale_price), LAND SQUARE FEET (land_square_feet), GROSS SQUARE FEET (gross_square_feet) sean de tipo flotante. Escribe quÃ© tipo de dato tienen asignado. Si no son correctos, realiza el cambio.

```python
target_cols = ["sale_price", "land_square_feet", "gross_square_feet"]

for col in target_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").astype(float)
```

Paso 3: Imprime de nuevo los tipos de datos para demostrar que has arreglado el problema.

```python
print(df.dtypes)
```

## PrÃ¡ctica 16: Genera el diccionario de datos del dataframe preprocesado. Imprime el contenido de dicho diccionario

```python
data_dictionary = {}

for col in df.columns:
    series = df[col]
    data_dictionary[col] = {
        "dtype": str(series.dtype),
        "non_null_count": int(series.notna().sum()),
        "null_count": int(series.isna().sum()),
        "example_values": series.dropna().unique()[:5].tolist()
    }

for col, info in data_dictionary.items():
    print(f"\nðŸ”¹ {col}")
    for k, v in info.items():
        print(f"  - {k}: {v}")
```
