Paso 1: Importar librerias

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Paso 2: Cargar dataset "penguins"

```python
df = sns.load_dataset("penguins")

print("Dataset 'penguins' cargado correctamente.")
print("Dimensiones:", df.shape)
df.head()
```

Paso 3: Verificar valores nulos

```python
print("Valores nulos por columna:")
print(df.isna().sum())
```

Paso 4: Limpiar dataset y mostrar nuevas dimensiones

```python
df = df.dropna()
print("Dataset limpiado. Nuevas dimensiones:", df.shape)
```

Paso 5: Se define un tema y una paleta de colores

```python
sns.set_theme(style="white")
color_pallet = "mako"
```

Paso 6: Usa plt.subplots() para crear una figura con 2 filas y 2 columnas.
Genera los siguientes gráficos:
* (0,0): sns.histplot(): Histograma de las longitudes de aletas.
* (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
* (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
* (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
* Personaliza los gráficos con títulos y colores adecuados.
* Ajusta el diseño con plt.tight_layout().

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# (0,0): sns.histplot(): Histograma de las longitudes de aletas.
sns.histplot(
    data=df,
    x="flipper_length_mm",
    hue="species",
    bins=20,
    palette=color_pallet,
    ax=ax[0, 0]
)

ax[0, 0].set_title("Histograma de la longitud de aletas")
ax[0, 0].set_xlabel("Longitud de aleta (mm)")
ax[0, 0].set_ylabel("Frecuencia")

# (0,1): sns.boxplot(): Boxplot comparando la masa corporal por especie.
sns.boxplot(
    data=df,
    x="species",
    y="body_mass_g",
    hue="species",
    palette=color_pallet,
    legend=False,
    ax=ax[0, 1]
)

ax[0, 1].set_title("Masa corporal por especie")
ax[0, 1].set_xlabel("Especie")
ax[0, 1].set_ylabel("Masa corporal (g)")

# (1,0): sns.scatterplot(): Dispersión entre masa corporal y largo de aletas.
sns.scatterplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    palette=color_pallet,
    ax=ax[1, 0]
)

ax[1, 0].set_title("Relacion entre masa corporal y longitud de aleta")
ax[1, 0].set_xlabel("Longitud de aleta (mm)")
ax[1, 0].set_ylabel("Masa corporal (g)")

# (1,1): sns.violinplot(): Violin plot de la masa corporal según la especie.
sns.violinplot(
    data=df,
    x="species",
    y="body_mass_g",
    hue="species",
    palette=color_pallet,
    legend=False,
    ax=ax[1, 1]
)


ax[1, 1].set_title("Distribucion de masa corporal por especie (Violin)")
ax[1, 1].set_xlabel("Especie")
ax[1, 1].set_ylabel("Masa corporal (g)")

plt.tight_layout()
plt.show()
```
