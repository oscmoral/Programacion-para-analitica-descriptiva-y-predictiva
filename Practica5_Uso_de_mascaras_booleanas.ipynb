**Practica: Uso de máscaras booleanas**

Objetivo: Crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50. Luego:

1.   Selecciona los elementos mayores que 25 utilizando una máscara booleana.
2.   Cuenta cuántos elementos son mayores que 25.

**Nota: Para esta practica trabajaremos directamente con la libreria NumPy**

Paso 1: Importar librarias random para trabajar con valores aleatorios y numpy para trabajar con arreglos

```python
import random
import numpy as np
```

Paso 2: Creamos el arreglo unidimensional con 15 numeros aleatorios entre 1 y 50

```python
array = np.random.randint(1, 51, 15)

print("Arreglo original:")
print(array)
```

Paso 3: Creamos la mascara booleana

```python
mask = array > 25

filtered_values = array[mask]
count_greater_25 = filtered_values.size
```

Paso 4: Imprimimos los valores

```python
print("\nMáscara booleana (mayores que 25):")
print(mask)

print("\nElementos mayores que 25:")
print(filtered_values)

print("\nCantidad de elementos mayores que 25:")
print(count_greater_25)
```
