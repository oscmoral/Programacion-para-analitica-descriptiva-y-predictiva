## Ejercicio 01: Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration

* Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.
* Cuenta la cantidad de registros duplicados utilizando solo la información de columna Duration e imprime
* Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.
* Filtra los datos y obten los registros sin duplicados

Paso 1: Importar librerias

```python
import pandas as pd
```

Paso 2: Acceso a Google Drive

```python
from google.colab import drive
drive.mount('/content/drive')
```

Paso 3: Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.

```python
path = "/content/drive/MyDrive/ClassFiles/dirtydata.csv"
df = pd.read_csv(path)

df.head()
```

Paso 4: Cuenta la cantidad de registros duplicados utilizando solo la información de columna Duration e imprime

```python
duplicados_duration = df.duplicated(subset=['Duration']).sum()

print("Registros duplicados en Duration:", duplicados_duration)
```

Paso 5: Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.

```python
conteo_duration = df['Duration'].value_counts()
print(conteo_duration)
```

Paso 6: Filtra los datos y obten los registros sin duplicados

```python
df_sin_duplicados_duration = df.drop_duplicates(subset=['Duration'])
df_sin_duplicados_duration
```
