## Ejercicio 02: Extraer componentes de fecha y tiempo

* De la columna 'DateTime' extrae el año, mes, día y hora. Crea una nueva columna para cada componente que extraes
* Imprime las primeras tres líneas del dataframe que ahora deberá tener 4 nuevas columnas.

Paso 1: De la columna 'DateTime' extrae el año, mes, día y hora. Crea una nueva columna para cada componente que extraes

```python
df["year"] = df["DateTime"].dt.year
df["month"] = df["DateTime"].dt.month
df["day"] = df["DateTime"].dt.day
df["hour"] = df["DateTime"].dt.hour
```

## Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)

* Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'.
* Guarda el resultado en una nueva columna y agrega al dataframe.
* Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.

Paso 1: Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'.

```python
delta = pd.Timedelta(days=10, hours=7, minutes=15)
```

Paso 2: Guarda el resultado en una nueva columna y agrega al dataframe.

```python
df["New_DateTime"] = df["DateTime"] + delta
```

Paso 3: Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.

```python
df[["DateTime", "New_DateTime"]].head(5)
```
