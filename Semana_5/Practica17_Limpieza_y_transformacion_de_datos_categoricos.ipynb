## Ejercicio 02: Errores Tipográficos y Valores Inconsistentes: highestdegree.

* Imprime de la columna 'highestdegrees' los valores únicos.
* Imprime la frecuencia de cada valor único
* Observa que cada valor categórico empieza con 0. Modifica las categorías para que todas estén en minúsculas y ninguna empiece con un número. Por ejemplo, si dice "2. High School", debería ser 'high school.
* Los cambios se realizarán en la misma columna 'highestdegrees'.
* Imprime los nuevos valores únicos.

Paso 1: Imprime de la columna 'highestdegrees' los valores únicos.

```python
print(df["highestdegree"].dropna().unique())
```

Paso 2: Imprime la frecuencia de cada valor único

```python
print(df["highestdegree"].value_counts(dropna=False))
```

Paso 3: Observa que cada valor categórico empieza con 0. Modifica las categorías para que todas estén en minúsculas y ninguna empiece con un número. Por ejemplo, si dice "2. High School", debería ser 'high school (Los cambios se realizarán en la misma columna 'highestdegrees').

```python
df["highestdegree"] = (
    df["highestdegree"]
    .astype("string")
    .str.replace(r"^\s*\d+\.\s*", "", regex=True)
    .str.strip()
    .str.lower()
)
```

Paso 4: Imprime los nuevos valores únicos.

```python
print(df["highestdegree"].dropna().unique())
```

## Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas

* De la columna 'maritalstatus' une las categorías Widowed y 'Single' en una categoría más general llamada 'Single/No Partner'. Crea una nueva columna, 'marital_grouped' donde aparezcan las nuevas categorías de maritalstatus.
* Imprime los valores únicos de la nueva columna 'marital_grouped'.
* Imprime la frecuencia de los valores únicos de la nueva columna 'marital_grouped'

Paso 1: De la columna 'maritalstatus' une las categorías Widowed y 'Single' en una categoría más general llamada 'Single/No Partner'. Crea una nueva columna, 'marital_grouped' donde aparezcan las nuevas categorías de maritalstatus.

```python
df["marital_grouped"] = df["maritalstatus"].replace({
    "Single": "Single/No Partner",
    "Widowed": "Single/No Partner"
})
```

Paso 2: Imprime los valores únicos de la nueva columna 'marital_grouped'.

```python
print(df["marital_grouped"].dropna().unique())
```

Paso 3: Imprime la frecuencia de los valores únicos de la nueva columna 'marital_grouped'

```python
print(df["marital_grouped"].value_counts(dropna=False))
```
