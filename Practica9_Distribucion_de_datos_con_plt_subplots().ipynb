**Practica: Distribución de Datos con plt.subplots()**

Objetivo:
Utilizar plt.subplots() para visualizar diferentes distribuciones generadas con NumPy y almacenadas en un DataFrame.

* Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).
* Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.
* Grafica cada distribución en un subplot usando ax[i, j].hist().
* Agrega títulos a cada subplot y ajusta la disposición con tight_layout().

Paso 1: Importar librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Paso 2: Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).

```python
np.random.seed(42)

size = 1000

data = {
    "Normal": np.random.normal(loc=0, scale=1, size=size),
    "Uniforme": np.random.uniform(low=0, high=1, size=size),
    "Exponencial": np.random.exponential(scale=1, size=size),
    "Poisson": np.random.poisson(lam=3, size=size)
}

df = pd.DataFrame(data)

df.head()
```

Paso 4: Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2, graficar cada distribución en un subplot usando ax[i, j].hist() y por ultimo y agregar títulos a cada subplot y ajusta la disposición con tight_layout().

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].hist(df["Normal"], bins=30)
ax[0, 0].set_title("Distribucion Normal")

ax[0, 1].hist(df["Uniforme"], bins=30)
ax[0, 1].set_title("Distribucion Uniforme")

ax[1, 0].hist(df["Exponencial"], bins=30)
ax[1, 0].set_title("Distribucion Exponencial")

ax[1, 1].hist(df["Poisson"], bins=30)
ax[1, 1].set_title("Distribucion Poisson")

plt.tight_layout()
plt.show()
```
