**Practica: Creación de un arreglo y operaciones básicas**

Objetivo: Crea un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100. Luego, realiza las siguientes operaciones:

1.   Encuentra el valor máximo del arreglo.
2.   Encuentra el valor mínimo del arreglo.
3.   Calcula la media del arreglo.
4.   Suma 10 a cada elemento del arreglo.

Paso 1: Para manejar elementos aleatorios importamos la libreria "random"

```python
import random
```

Paso 2: Creamos el arreglo con valores aleatorios

```python
array = [random.randint(1, 100) for _ in range(20)]

print("Arreglo original:")
print(array)
```

Paso 3: Realizamos las operaciones correspondientes

```python
# Calcular valor máximo
max_value = max(array)

# Calcular valor mínimo
min_value = min(array)

# Calcular la media
mean_value = sum(array) / len(array)
```

Paso 4: Imprimimos los valores obtenidos de las operaciones

```python
print("\nResultados:")
print("Valor máximo:", max_value)
print("Valor mínimo:", min_value)
print("Media:", mean_value)
```

Paso 5: Sumamos 10 a cada elemento del arreglo

```python
array_plus_10 = [value + 10 for value in array]

print("\nArreglo con 10 sumado a cada elemento:")
print(array_plus_10)
```
