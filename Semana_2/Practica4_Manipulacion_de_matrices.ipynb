**Practica: Manipulación de matrices**

Objetivo: Crea dos matrices de 3x3 con números aleatorios entre 1 y 10. Realiza las siguientes operaciones:

1.   Suma ambas matrices.
2.   Resta la segunda matriz de la primera.
3.   Multiplica la primera matriz por 2.
4.   Realiza una multiplicación matricial entre las dos matrices.

Paso 1: Para manejar elementos aleatorios importamos la libreria "random"

```python
import random
```

Paso 2: Creamos las dos matrices 3x3 con valores aleatorios utilizando listas anidadas

```python
matrix_a = [[random.randint(1, 10) for _ in range(3)] for _ in range(3)]
matrix_b = [[random.randint(1, 10) for _ in range(3)] for _ in range(3)]

print("Matriz A:")
for row in matrix_a:
    print(row)

print("\nMatriz B:")
for row in matrix_b:
    print(row)
```

Paso 3: Sumamos las matrices

```python
sum_matrix = [
    [matrix_a[i][j] + matrix_b[i][j] for j in range(3)]
    for i in range(3)
]

print("\nSuma de A + B:")
for row in sum_matrix:
    print(row)
```

Paso 4: Restamos las matrices

```python
sub_matrix = [
    [matrix_a[i][j] - matrix_b[i][j] for j in range(3)]
    for i in range(3)
]

print("\nResta de A - B:")
for row in sub_matrix:
    print(row)
```

Paso 5: Multiplicamos la matriz A por 2

```python
scalar_matrix = [
    [value * 2 for value in row]
    for row in matrix_a
]

print("\nMatriz A multiplicada por 2:")
for row in scalar_matrix:
    print(row)
```

Paso 6: Multiplicamos la matriz (A x B)

```python
mul_matrix = [[0 for _ in range(3)] for _ in range(3)]

for i in range(3):
    for j in range(3):
        for k in range(3):
            mul_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]

print("\nMultiplicación matricial A x B:")
for row in mul_matrix:
    print(row)
```

Paso 1: importamos la libreria NumPy

```python
matrix_a = np.random.randint(1, 11, (3, 3))
matrix_b = np.random.randint(1, 11, (3, 3))

print("Matriz A:")
print(matrix_a)

print("Matriz B:")
print(matrix_b)
```

Paso 3: Las operaciones podrian quedar reducidas a lo siguiente:

```python
print("Suma:\n", matrix_a + matrix_b)
print("Resta:\n", matrix_a - matrix_b)
print("Escalar x2:\n", matrix_a * 2)
print("Multiplicación matricial:\n", matrix_a @ matrix_b)
```
